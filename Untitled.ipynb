```python
import matplotlib.pyplot

# define subplots
fig, ax = matplotlib.pyplot.subplots(2, 1, gridspec_kw={'height_ratios': [2, 1]})
# define data
x = [0, 1, 2, 3, 4, 5]
y = [i**2 for i in x]
# create subplots
ax[0].plot(x, y, color='black')
ax[1].plot(x, y, color='green')
# display the plot
matplotlib.pyplot.show()
```
